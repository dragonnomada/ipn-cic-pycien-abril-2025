```python
data1 = titanic[["Sex", "Survived"]]

data1
```

```python
report1 = data1.groupby("Sex").aggregate("mean")
report1
```

```python
seaborn.countplot(report1)
```

```python
import scipy.stats as stats

x1 = titanic[(titanic["Sex"] == "male") & (titanic["Pclass"] == 1)]["Survived"]
x2 = titanic[(titanic["Sex"] == "male") & (titanic["Pclass"] == 3)]["Survived"]

t_stat, p_val = stats.ttest_ind(x1, x2)

print(f"T-Test: {t_stat:.2f}, Valor-P: {p_val:.2f}")
```
